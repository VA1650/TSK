```python
import pandas as pd

#Файл переводим в кодировку UTF-8

data = pd.read_csv('town_1959.csv') #Считаем данные с файла

print(data.head()) #Проверим корректность импорта данных

print(len(data)) #Определим количество строк

min_population_city = data.loc[data['население'].idxmin()]['город']
print("Минимальное значение населения:", min_population_city) #Вычислим минимальное значение используя встроенные средства инструмента анализа

min_population = float('inf')
min_population_city = ''
for index, row in data.iterrows():
  if row['население'] < min_population:
    min_population = row['население']
    min_population_city = row['город']
print("Минимальное значение населения:", min_population_city) #Вычислим минимальное значение используя цикл

max_population_city = data.loc[data['население'].idxmax()]['город']
print("Максимальное значение населения:", max_population_city) #Вычислим максимальное значение используя встроенные средства инструмента анализа

max_population = float('-inf')
max_population_city = ''
for index, row in data.iterrows():
  if row['население'] > max_population:
    max_population = row['население']
    max_population_city = row['город']
print("Максимальное значение населения:", max_population_city) #Вычислим максимальное значение используя цикл

population_mean_builtin = data['население'].mean()
print("Среднее значение населения:", population_mean_builtin) #Вычислим среднее значение используя встроенные средства инструмента анализа

total_population = 0
for index, row in data.iterrows():
  total_population += row['население']
avg_population = total_population / len(data)
print("Среднее значение населения:", avg_population) #Вычислим среднее значение используя цикл

import matplotlib.pyplot as plt

data.plot(y='население', kind='bar', figsize=(10, 6))
plt.show() #Построим столбчатую диаграмму количество городов от значения населения

import numpy as np

plt.figure(figsize=(10, 6))
plt.bar(data['город'], np.log(data['население']))
plt.show() #Построим столбчатую диаграмму логарифмического распределения

top_cities = data.head(8)

plt.figure(figsize=(8, 8))
plt.pie(top_cities['население'], labels=top_cities['город'], autopct='%1.1f%%')
plt.show() #Построим круговую диаграмму количество городов от значения населения


plt.figure(figsize=(12, 6))
plt.plot(data['город'], data['население'], label='Population')
plt.bar(data['город'], data['население'], label='Population')
plt.legend()
plt.show()#Представим два любых графика в одном окне вывода
```
